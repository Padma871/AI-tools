#  Matplotlib

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style #adding style
```

# Linear graph

```python
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.plot(x,y,label="linear line",color='g')
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# scatter plot

```python
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.scatter(x,y,label="scatter plot",color='g',marker='x')
plt.title("sample scatter graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True,color='k')
plt.legend()
plt.show()
```
